```python
'''
def sample_policy(observation):
    """
    A policy that sticks if the player score is >= 20 and hits otherwise.
    """
    score, dealer_score, usable_ace = observation
    
    if score >= 17:
        action = 0
    elif ( score < 17 and score>=13 )  and ( dealer_score==2 or dealer_score==3 or dealer_score==4 or dealer_score==5 or dealer_score==6) :
        action = 0
    elif score == 12 and ( dealer_score == 4 or dealer_score == 5 or dealer_score == 6):
        action = 0
    else:
        action = 1
    return action
'''

def sample_policy(observation):
    """
    A policy that sticks if the player score is >= 20 and hits otherwise.
    """
    score, dealer_score, usable_ace = observation
    return 0 if score >= 12 else 1


def mc_prediction(policy, environment, num_episodes, discount_factor=1.0):
    """
    Monte Carlo prediction algorithm. Calculates the value function
    for a given policy using sampling.
    
    Args:
        policy: A function that maps an observation to action probabilities.
        env: OpenAI gym environment.
        num_episodes: Number of episodes to sample.
        discount_factor: Gamma discount factor.
    
    Returns:
        A dictionary that maps from state -> value.
        The state is a tuple and the value is a float.
    """

    # Keeps track of sum and count of returns for each state
    # to calculate an average. We could use an array to save all
    # returns (like in the book) but that's memory inefficient.
    returns_sum = defaultdict(lambda: "Not Present")
    #print("returns_sum",returns_sum[:])
    returns_count = defaultdict(lambda:0)
    #print("returns_count",returns_count[:])
    
    # The final value function
    V = defaultdict(lambda: "Not Present")
    #print("V=",V)
    Q = defaultdict(lambda: "Not Present")


    for i in range(1,num_episodes+1):
        # 1. get the first state of environment
        #    The observation of a 3-tuple of: the players current sum,
        #    the dealer's one showing card (1-10 where 1 is ace),
        #    and whether or not the player holds a usable ace (0 or 1).
        #    state= (17, 6, True)
        #    state= (14, 2, False)
        #    state= (14, 2, False)  ......

        if i%1000 == 0:
            print(i)
        state = environment.reset()
        #print("state=",state)
        done = False
        #print("V[state]=",V[state])
        episode = []
        
        # one opisode game max length is 5
        for i in range(100):
        # 2. get the action by the policy
            if not done:
                action = policy(state)
        # 3. input the action into the environment
                next_state, reward, done, _ = environment.step(action)
                #print("next_state=",state)
                #print("reward=",reward)
                #print("done=",done)
                episode.append([state,action,reward])
                #print("episode=",episode)
                #print("i",i)
                state = next_state
                if done:
                    break
        #print("=================================")
        for index, mdp in enumerate(episode):
            #print("index=", index)
            #print("mdp=",mdp) #, mdp= [(12, 5, False), 0, -1]

            if V[mdp[0]] =="Not Present":
                #print("state is not used, state = ", mdp[0])
                Gt = []
                for x in range(index, len(episode)):
                    Gt.append(episode[x][2])
                returns_sum[mdp[0]] = sum(Gt[:])
                returns_count[mdp[0]] = 1;
                V[mdp[0]] = returns_sum[mdp[0]]/returns_count[mdp[0]]
            else:
                Gt = []
                for x in range(index, len(episode)):
                    Gt.append(episode[x][2])
                    #print("Gt=",Gt)
                    
                
                returns_sum[mdp[0]] += sum(Gt[:])
                returns_count[mdp[0]] += 1;
                V[mdp[0]] = returns_sum[mdp[0]]/returns_count[mdp[0]]
        #if not done :
            
        # 4. get the observation from the environment
        # 5. use the default policy to run MC search to the end of the episode
        # 6. backup the trajectory of each episode
        # 7. use each episode to update the Value function of each state in the game
        
        for index, mdp in enumerate(episode):
            #print("index=", index)
            #print("mdp=",mdp)

            if Q[(mdp[0],mdp[1])] =="Not Present":
                #print("state is not used, state = ", mdp[0])
                Gt = []
                for x in range(index, len(episode)):
                    Gt.append(episode[x][2])
                returns_sum[(mdp[0],mdp[1])] = sum(Gt[:])
                returns_count[(mdp[0],mdp[1])] = 1;
                Q[(mdp[0],mdp[1])] = returns_sum[(mdp[0],mdp[1])]/returns_count[(mdp[0],mdp[1])]
            else:
                Gt = []
                for x in range(index, len(episode)):
                    Gt.append(episode[x][2])
                    #print("Gt=",Gt)
                    
                
                returns_sum[(mdp[0],mdp[1])] += sum(Gt[:])
                returns_count[(mdp[0],mdp[1])] += 1;
                Q[(mdp[0],mdp[1])] = returns_sum[(mdp[0],mdp[1])]/returns_count[(mdp[0],mdp[1])]
        
        #if not done :
            
        # 4. get the observation from the environment
        # 5. use the default policy to run MC search to the end of the episode
        # 6. backup the trajectory of each episode
        # 7. use each episode to update the Value function of each state in the game
    for key,value in V.items():
          print (key,value)    
    for key,value in Q.items():
          print (key,value)  
    return V, Q
```

```python
%matplotlib inline

import gym
import matplotlib
import numpy as np
import sys

from collections import defaultdict

if "../" not in sys.path:
  sys.path.append("../") 
from lib.envs.blackjack import BlackjackEnv
from lib import plotting

matplotlib.style.use('ggplot')
```

```python
env = BlackjackEnv()
```

```python
def mc_prediction(policy, env, num_episodes, discount_factor=1.0):
    """
    Monte Carlo prediction algorithm. Calculates the value function
    for a given policy using sampling.
    
    Args:
        policy: A function that maps an observation to action probabilities.
        env: OpenAI gym environment.
        num_episodes: Number of episodes to sample.
        discount_factor: Gamma discount factor.
    
    Returns:
        A dictionary that maps from state -> value.
        The state is a tuple and the value is a float.
    """

    # Keeps track of sum and count of returns for each state
    # to calculate an average. We could use an array to save all
    # returns (like in the book) but that's memory inefficient.
    returns_sum = defaultdict(float)
    returns_count = defaultdict(float)
    
    # The final value function
    V = defaultdict(float)
    
    for i_episode in range(1, num_episodes + 1):
        # Print out which episode we're on, useful for debugging.
        if i_episode % 1000 == 0:
            print("\rEpisode {}/{}.".format(i_episode, num_episodes), end="")
            sys.stdout.flush()

        # Generate an episode.
        # An episode is an array of (state, action, reward) tuples
        episode = []
        state = env.reset()
        for t in range(100):
            action = policy(state)
            next_state, reward, done, _ = env.step(action)
            episode.append((state, action, reward))
            if done:
                break
            state = next_state
        print("episode length",len(episode))
        # Find all states the we've visited in this episode
        # We convert each state to a tuple so that we can use it as a dict key
        states_in_episode = set([tuple(x[0]) for x in episode])
        for state in states_in_episode:
            # Find the first occurance of the state in the episode
            first_occurence_idx = next(i for i,x in enumerate(episode) if x[0] == state)
            # Sum up all rewards since the first occurance
            G = sum([x[2]*(discount_factor**i) for i,x in enumerate(episode[first_occurence_idx:])])
            # Calculate average return for this state over all sampled episodes
            returns_sum[state] += G
            returns_count[state] += 1.0
            V[state] = returns_sum[state] / returns_count[state]

    return V
```

```python
def sample_policy(observation):
    """
    A policy that sticks if the player score is >= 20 and hits otherwise.
    """
    score, dealer_score, usable_ace = observation
    return 0 if score >= 19 else 1

V_10k = mc_prediction(sample_policy, env, num_episodes=1000000)
plotting.plot_value_function(V_10k, title="10,000 Steps")
```
